```python
# Find statistics about the Price and Time column

# Price mean, std dev, min, max, range, variance
price_mean = mc_data['Price'].mean()
price_std = mc_data['Price'].std()
price_min = mc_data['Price'].min()
price_max = mc_data['Price'].max()
price_range = price_max - price_min
price_variance = numpy.var(mc_data['Price'])

# Time mean, std dev, min, max, range, variance
time_mean = mc_data['Time'].mean()
time_std = mc_data['Time'].std()
time_min = mc_data['Time'].min()
time_max = mc_data['Time'].max()
time_range = time_max - time_min
time_variance = numpy.var(mc_data['Time'])


# Print the statistics
print('Price mean:', price_mean)
print('Price std dev:', price_std)
print('Price min:', price_min)
print('Price max:', price_max)
print('Price range:', price_range)
print('Price variance:', price_variance)

print('Time mean:', time_mean)
print('Time std dev:', time_std)
print('Time min:', time_min)
print('Time max:', time_max)
print('Time range:', time_range)
print('Time variance:', time_variance)
```
