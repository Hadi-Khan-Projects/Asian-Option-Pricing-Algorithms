### Plotting the Difference between Gaussian and each Levy Process

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
```

```python
# Get the data
data = pd.read_csv('Levy_process_difference.csv')

# Split data based on the type of the process
data_nig = data[data['Process'] == 'NIG'].reset_index()
data_cgmy = data[data['Process'] == 'CGMY'].reset_index()
data_de = data[data['Process'] == 'DE'].reset_index()
data_jd = data[data['Process'] == 'JD'].reset_index()
```

```python
colors = ['navy', 'darkorange', 'forestgreen', 'crimson']

# Interpolating the data for each process
X_new = np.linspace(80, 120, 1000)  # Smooth curve

# JD Process (existing data)
spl_jd = make_interp_spline(data_jd['Strike'], data_jd['Pct_Difference'], k=3)
Y_jd_smooth = spl_jd(X_new)

# NIG Process
spl_nig = make_interp_spline(data_nig['Strike'], data_nig['Pct_Difference'], k=3)
Y_nig_smooth = spl_nig(X_new)

# CGMY Process
spl_cgmy = make_interp_spline(data_cgmy['Strike'], data_cgmy['Pct_Difference'], k=3)
Y_cgmy_smooth = spl_cgmy(X_new)

# DE Process
spl_de = make_interp_spline(data_de['Strike'], data_de['Pct_Difference'], k=3)
Y_de_smooth = spl_de(X_new)

# Plotting
plt.figure(figsize=(12, 8), dpi=800)
plt.plot(X_new, Y_jd_smooth, label='jd vs. g', color=colors[0])
plt.plot(X_new, Y_nig_smooth, label='nig vs. g', color=colors[1])
plt.plot(X_new, Y_cgmy_smooth, label='cgmy vs. g', color=colors[2])
plt.plot(X_new, Y_de_smooth, label='de vs. g', color=colors[3])
plt.margins(x=0)
plt.xlabel('Strike Price')
plt.ylabel('Price Difference (Percentage)')
plt.ylim(-100, 100)  # Y-axis limits
plt.grid(True)
plt.legend()
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)
plt.savefig("PLOT_Levy_Process_Difference.png", bbox_inches='tight', dpi=800)
plt.show()
```
